```python
class Candidate:
    def __init__(self, name):
        self.name = name
        self.vote_count = 0

    def add_vote(self):
        self.vote_count += 1


class VotingSystem:
    def __init__(self):
        self.candidates = []

    def add_candidate(self, name):
        candidate = Candidate(name)
        self.candidates.append(candidate)

    def show_candidates(self):
        print("\nCandidates:")
        for i, candidate in enumerate(self.candidates, start=1):
            print(f"{i}. {candidate.name}")

    def vote(self):
        self.show_candidates()
        try:
            choice = int(input("Enter candidate number to vote: "))
            if 1 <= choice <= len(self.candidates):
                self.candidates[choice - 1].add_vote()
                print("Vote recorded successfully!\n")
            else:
                print("Invalid choice.\n")
        except ValueError:
            print("Please enter a valid number.\n")

    def show_results(self):
        print("\nVoting Results:")
        for candidate in self.candidates:
            print(f"{candidate.name}: {candidate.vote_count} votes")


# Main program
voting_system = VotingSystem()

# Add candidates
voting_system.add_candidate("Alice")
voting_system.add_candidate("Bob")
voting_system.add_candidate("Charlie")

# Voting loop
while True:
    print("\n1. Vote")
    print("2. Show Results")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        voting_system.vote()
    elif choice == "2":
        voting_system.show_results()
    elif choice == "3":
        print("Exiting voting system.")
        break
    else:
        print("Invalid choice. Try again.")
```
