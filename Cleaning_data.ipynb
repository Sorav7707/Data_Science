```python
df = df.dropna(how = 'all')
# that query remove full row is it empty
```

```python
df.isna().all(axis=1).sum()
# check all empty row remove
```

```python
bad_values = ['ERROR', 'UNKNOWN', 'Unknown', 'error', '', ' ']

cols = ['Item', 'Price Per Unit', 'Quantity','Total Spent','Discount Applied']

for col in cols:
    df[col] = df[col].replace(bad_values, pd.NA)
# Standardize (make all bad values uniform)
# Before recovery, convert everything bad into real NaN.
    
# âœ… Now pandas clearly knows what is missing
# (This is a mandatory professional step)
```

```python
df['Item'] = df.groupby('Category')['Item'].transform(
    lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x
)
#Recover data using logic (this is the core skill)
#ðŸ”¹1.Recover Item using Category (best beginner logic)
# If Category â†’ Item relationship exists:

# ðŸ§  Why companies like this:
# Category is stable
# No guessing
# Data-driven recovery
```

```python
df['Payment Method'] = df.groupby('Customer ID')['Payment Method'].transform(
    lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x
)
# 2.Recover Payment Method using Customer behavior
# Customers often use the same payment method repeatedly

# âœ” Very realistic business logic
#âœ” Used in retail & fintech datasets
```

```python
df['Location'] = df.groupby('Item')['Location'].transform(
    lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x
)
# 3. Recover Location using Item
# Some items are mostly:

# Online
# In-store
```

```python
df[['Item', 'Payment Method', 'Location']] = df[
    ['Item', 'Payment Method', 'Location']
].fillna('UNRESOLVED')
# 3ï¸âƒ£ Explicit labeling (never hide uncertainty)

# After all logical recovery, some values may still be missing.
# Companies want them clearly labeled, not silently filled.

# â€œIf you donâ€™t know it for sure, label it clearly.â€
```

```python
# Final Validation (company expects this)
df[['Item', 'Payment Method', 'Location']].isna().sum()
```
